Caso existam dados duplicados, remova-os. Caso existam dados ausentes, voc√™ pode remov√™-los ou pensar em alguma maneira de preench√™-los.

> [Clique aqui](https://s3-sa-east-1.amazonaws.com/lcpi/e1c2115b-1725-4525-9fd5-0ce57fb5c408.csv) para fazer o download do dataset.

````python
def show_nan_stats(df):
  print('>> VALORES ABSOLUTOS <<')
  print(df.isnull().sum(), '\n')
  
  print('>> VALORES PERCENTUAIS <<')
  print(df.isnull().sum() / df.shape[0] * 100)

google_play_store = pd.read_csv('googleplaystore.csv')
show_nan_stats(google_play_store)

google_play_store.drop_duplicates(ignore_index=True, inplace = True)
google_play_store.dropna()
````
